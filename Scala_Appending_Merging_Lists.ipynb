```python
!apt-get update
!apt-get install -y scala
!scala -version
#define scala code as a multi line string
scala_code="""
object NeuralNetwork {
  //define a layer case class to represent layers in the network
  case class Layer(name:String,units:Int,activation:Option[String],dropout:Option[Double])
  //Initial Layers:Input and Output
  val inputLayer=List (Layer("Input",784,None,None))
  var hiddenLayers=List(
    Layer("Dense_1",128,Some("relu"),Some(0.2)),
    Layer("Dense_2",64,Some("relu"),Some(0.3))
  )
  val outputLayer=List(Layer("Output",10,Some("softmax"),None))
  //Function to add a new layer
  def addLayer(layers: List[Layer],name:String,units:Int,activation:String,dropout: Option[Double]=None):List[Layer]={
    layers:+Layer(name,units,Some(activation),dropout)
  }
  def main(args:Array[String]):Unit={
    //Adding a new hidden layer
    hiddenLayers=addLayer(hiddenLayers,"Dense_3",32,"relu",Some(0.4))
    //Merging all layers into the final neural network architecture
    val neuralNetwork=inputLayer:::hiddenLayers:::outputLayer
   //Displaying the complete NN architecture
   neuralNetwork.foreach { layer =>
      println(s"Layer: ${layer.name},Units:${layer.units},Activation:${layer.activation.getOrElse("None")}")
      }
  }
  //Invoke the main method
  main(Array())
}
"""
#write the scala code to a file named NeuralNetwork.scala
with open("NeuralNetwork.scala","w") as file:
  file.write(scala_code)
#compile the scala code
!scalac NeuralNetwork.scala
#run the compiled Scala code
!scala NeuralNetwork
```
